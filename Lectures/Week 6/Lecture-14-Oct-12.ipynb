````python
from scipy.linalg import hilbert

def compare_hilbert_inverse(n, epsilon=1e-10):

    H = hilbert(n)
    H_perturbed = H + epsilon * np.random.randn(*H.shape)

    H_perturbed_inv = np.linalg.inv(H_perturbed)
    
    plt.figure(figsize=(12, 5))
    plt.subplot(121)

    plt.imshow(np.linalg.inv(H), cmap='viridis')
    plt.colorbar()
    plt.title("Inverse of H")
    plt.subplot(122)
    plt.imshow(H_perturbed_inv, cmap='viridis')
    plt.colorbar()
    plt.title("Inverse of Perturbed H")

    plt.show()
````

````python
compare_hilbert_inverse(10, epsilon=1e-20)
````

````python
compare_hilbert_inverse(10, epsilon=1e-10)
````

````python
compare_hilbert_inverse(50, epsilon=1e-2)
````
