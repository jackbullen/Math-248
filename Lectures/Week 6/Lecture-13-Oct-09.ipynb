Which points converge to which roots in newton's method?

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.array([x[0]**2 - x[1] + x[0]*np.sin(x[0]), x[0] + x[1]**2 - 3])
```

```python
from itertools import product
from collections import defaultdict

xs = np.linspace(-3, 3, 100)
ys = np.linspace(-3, 3, 100)

roots = defaultdict(list)

color = plt.cm.rainbow(np.linspace(0, 1, len(xs)*len(ys)))

for x, y in product(xs, ys):
    root = fsolve(f, [x, y])
    
    root = tuple(np.round(root, 3))
    try :
        roots[root].append((x, y))
        plt.plot(x, y, '.', color=color[len(roots[root])-1])
    except KeyError:
        roots[root] = [(x, y)]
        plt.plot(x, y, '.', color=color[0])
```
