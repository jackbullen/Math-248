### let r>1. want to find x such that 1-x = e^{-rx}

````python
r=2

f = lambda x: 1-x-np.exp(-r*x)

fp = lambda x: -1+r*np.exp(-r*x)

def zero_finder(x):
    return x-f(x)/fp(x) 

tol = 1e-08

root = 0.8
for i in range(100000):
    root = zero_finder(root) 
print(root)

xs = np.linspace(-0.25,1,1000)
ys = [f(x) for x in xs]
plt.grid()
plt.plot(xs, ys)
plt.axvline(root, c='r')
````


We look for the root of a function, $f(x)$, i.e., $f(x)=0$

### Use the derivative to narrow down on the root.
$$ y = f(x) $$
  $$ y-y_0 = m(x-x_0) $$
  $$ y-f(x_0) = f'(x_0)(x-x_0)$$
  $$ x_{n+1} = x_n - \frac {f(x_n)}{f'(x_n)}$$
