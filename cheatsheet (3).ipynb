### Root Finding

 $$ x_{n+1} = x_n - \frac {f(x_n)}{f'(x_n)}$$

```python
# With a for loop

r=2

f = lambda x: 1-x-np.exp(-r*x)

fp = lambda x: -1+r*np.exp(-r*x)

def zero_finder(x):
    return x-f(x)/fp(x) 

tol = 1e-08

root = 0.4
for i in range(100000):
    root = zero_finder(root) 
print(f"the root is {root}")
```

```python
# With a while loop

def f(x):
    return 1-x-np.exp(-r*x)

def fp(x):
    return -1+r*np.exp(-r*x)

new_try_at_root = 7
ct = 0

print('func val,', 'root,', 'iteration count')
while np.abs(f(new_try_at_root)) > tol:
    new_try_at_root = zero_finder(new_try_at_root)
    print(f"{f(new_try_at_root):.4f}, {new_try_at_root:.4f}, {ct}")
    ct+=1
```

```python
# basic plt.scatter() and plt.figure(): rotating a square lattice by pi/4

import itertools as it

A = np.matrix([[np.cos(np.pi/4), -np.sin(np.pi/4)],[np.cos(np.pi/4), np.sin(np.pi/4)]])

xs = np.linspace(-1,1,10)
ys = np.linspace(-1,1,10)

plt.figure()
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
for x,y in it.product(xs, ys):
    plt.scatter(x, y)
    

plt.figure()
for x,y in it.product(xs, ys):
    plt.scatter(A[0,0]*x+A[0,1]*y, A[1,0]*x+A[1,1]*y)

```

```python
import time

start_time = time.time()

# can easily make counting datasets with default dic
di = defaultdict(list)

for i in range(1, 10000):
    for j in range(2, i):
        if i%j==0:
            di[j].append(i)

end_time = time.time()

print(f"{end_time-start_time:.2f}")
```
