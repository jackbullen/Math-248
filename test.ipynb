```python
from itertools import product
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.sqrt(0.1)
```

```python
Z = list(range(0, 11))
Z
```

```python
for x in product(Z, Z):
    plt.plot(x[0], x[1], '.', c='r')
```

```python
lattice = list(product(Z, Z))
```

```python
for pt, pt2 in product(lattice, lattice):
    rand = np.random.rand()
    if abs(pt[0]-pt2[0]) + abs(pt[1]-pt2[1]) > 80*rand:

        plt.plot([pt[0], pt2[0]], [pt[1], pt2[1]], c='b', alpha=0.1)
```
