```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

```python
resolution = 1920
X = np.linspace(-10, 10, resolution)
Y = np.linspace(-10, 10, resolution)
```

```python
pixel_array = np.ndarray((resolution, resolution))
for i, j in product(range(resolution), range(resolution)):
    pixel_array[i, j] = np.sin(X[i] * Y[j])
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(pixel_array, cmap='gray')
plt.show()
```

```python
def cplot(f, resolution=1000, extent=[-1, 1, -1, 1], cmap='magma', vmin=None, vmax=None):
    X = np.linspace(-10, 10, resolution)
    Y = np.linspace(-10, 10, resolution)

    pixel_array = np.ndarray((resolution, resolution), dtype=float)

    for i, j in product(range(resolution), range(resolution)):
        pixel_array[i, j] = f(complex(X[i], Y[j])).real

    plt.figure(figsize=(8, 8))
    plt.imshow(pixel_array, extent = extent, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.colorbar()
    plt.show()

    return pixel_array
```

```python
cplot(lambda z: (z**2+2)/(z**2-4), vmin=-1, vmax=1, cmap='seismic')
```

```python
cplot(lambda z: (z**2-2)/(z**3+2) + 1/z**12 + (z**2-0.1)/(z**2+12), extent=[-0.7, 0.7, -0.7, 0.7], vmin=-0.22, vmax=2)
```
