```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import fsolve
```

$$f(x) = y$$
$$x = f^{-1}(y)$$
How can we solve for $f^{-1}$?

- We know $f(x)-y=0$. 

- We can find the roots using `fsolve()`.

Given f and y, find $x$ such that  

$$f(x) - y = 0$$

```python
# inv = lambda f, y: fsolve(lambda x: f(x)-y, 1)[0]

def inv(f,y):
    if type(y) is not np.ndarray:
        print(type(y))
        return fsolve(lambda x: f(x)-y, 1)[0]
    return np.array([fsolve(lambda x: f(x)-yi, 1)[0] for yi in y])
```

```python
np.exp(1), inv(np.exp, 1), np.log(1)
```

```python
# be careful about when the function is not invertible
np.exp(-2), inv(np.exp, -2), np.log(-2)
```

```python
xs = np.linspace(-1, 1, 100)
xs_inv = np.linspace(1e-2, 2, 100)
plt.plot(xs_inv, inv(np.exp, xs_inv))
plt.plot(xs, xs)
plt.plot(xs, np.exp(xs))
plt.axis('equal')
```

```python
eek = lambda x: x**x

xs = np.linspace(-1, 2, 100)

eek_inv = inv(eek, xs)
ys = eek(xs)

plt.plot(xs, ys)
plt.plot(xs, xs)
plt.plot(xs, eek_inv)
plt.axis('equal')
```
