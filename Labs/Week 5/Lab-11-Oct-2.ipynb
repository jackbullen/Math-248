```python
import numpy as np
import matplotlib.pyplot as plt
```

# $ x_{n+1} = x_n e^{r(1-\frac{x_n}{K})}$

```python
r = 2
K = 1000

def ricker_model(x):
    return x * np.exp(r * (1 - x / K))
```

```python
def run(f, x0, n=1000):
    xs = [x0]
    for i in range(n):
        xs.append(f(xs[-1]))
    return xs
```

```python
xs = run(ricker_model, 1, 100)
plt.plot(xs)
```

```python
plt.figure(figsize=(12,12))
for r in np.linspace(0, 4, 10000):
    xs = run(ricker_model, 0.1, 100)[-20:]
    plt.plot([r]*len(xs), xs, '.', markersize=0.7)
plt.xlim(0,4)
```

```python
plt.figure(figsize=(12,12))
for r in np.linspace(3.6, 3.95, 15000):
    xs = run(ricker_model, 0.1, 100)[-20:]
    plt.plot([r]*len(xs), xs, '.', markersize=0.5)
```
