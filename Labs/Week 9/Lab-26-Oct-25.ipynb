```python
import numpy as np
from numpy.random import binomial as B

import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
def walk(p=0.5, n=100):
    trials = np.array(B(1, p, n))*2 - 1
    xs = [sum(trials[:i]) for i in range(len(trials))]
    return xs
```

```python
nwalks = 36000
nsteps = 100

walks = []
for _ in range(nwalks):
    xs = walk(n=nsteps)
    walks.append(xs)
    plt.plot(xs)
```

```python
returns = defaultdict(int)

for xs in walks:
    for i, loc in enumerate(xs[1:]):
        if loc == 0:
            returns[i+1] += 1
            break
```

```python
tally = 0
for key, val in sorted(returns.items(), key=lambda x: x[0]):
    print(key, f'{val/nwalks: 0.8f}', val, sep='\t')
    tally += val
print()
print(f'{nwalks-tally} out of {nwalks} did not return to 0 in {nsteps} steps.')
```
