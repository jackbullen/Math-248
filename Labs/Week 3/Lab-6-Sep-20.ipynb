```python
theta = np.pi/4
rotate = np.matrix([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])

print("M =", rotate, end="\n\n")
print("(row, col) =", rotate.shape, end="\n\n")
print("det(M) =", np.linalg.det(rotate), end="\n\n")
print("norm(M)", np.linalg.norm(rotate), end="\n\n")
print("cond(M)", np.linalg.cond(rotate), end="\n\n")

print("Mx = <root(2)/2, root(2)/2>. x = <1,0> =", np.linalg.solve(rotate, np.array([np.sqrt(2)/2, np.sqrt(2)/2])), end="\n\n")
print("M^-1 =", rotate.I, end="\n\n")
print("M*M^-1 =", rotate*rotate.I, end="\n\n")
print("M^T =", rotate.T, end="\n\n")

print("eigenvalues of M =", np.linalg.eigvals(rotate), end="\n\n")
print("eigenvectors of M =", np.linalg.eig(rotate), end="\n\n")

print("qr decomposition of M =", np.linalg.qr(rotate), end="\n\n")
```

Consider the vectors $\vec{v} = <1,2,3>$ and $\vec{u} = <1,1,-1>$



```python
u = np.array([1,1,-1])
v = np.array([1,2,3])
```

Their dot product is

$$ u \cdot v = |u||v|cos(\theta) $$

Therefore

$$ u \cdot v = 1*1 + 2*1 + 3*-1 = 0 $$

```python
np.matmul(u,v)
```

Angle between them is

$$ \theta = \arccos{\frac{u \cdot v}{\|u\|\|v\|}} = \arccos 0 = \frac{\pi}{2} $$

```python
np.arccos(np.matmul(u,v)/(np.linalg.norm(u)*np.linalg.norm(v)))
```

## Fibonacci Sequence

```python
n = 10

fibs = [ (((1+np.sqrt(5))/2)**n - ((1-np.sqrt(5))/2)**n) / np.sqrt(5) for n in range(n)]

ratios = [fibs[i] / fibs[i-1] for i in range(1, n)]

fibs[-1], ratios[-1]
```

```python
# with recursion, but not using already computed values

def fibonacci(n, f0=0, f1=1):
    if n>1:
        f1, f0 = fibonacci(n-1, f0, f1)
        return f1 + f0, f0
    elif n==1:
        return f1, f0
    else:
        return f0, None

def fib(n):
    return [fibonacci(i) for i in range(0, n)]

start = time.time()
fib(1000)
end = time.time()
print(f'\nRunning time: {end - start:.6f} seconds')
```

```python
# with a loop, dynamic

def fib2(n):
    nums = [0, 1]
    for i in range(2, n+1):
        nums.append(nums[i-1] + nums[i-2])
    return nums

start = time.time()
fib2(1000)
end = time.time()
print(f'\nRunning time: {end - start:.6f} seconds')
```
