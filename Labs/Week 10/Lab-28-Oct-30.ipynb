```python
# sampling from a normal distribution to get planar brownian motion
# 2D random walk

X = np.random.normal(0,1,2000)
X = np.cumsum(X)
plt.plot(X)
```

```python
def brownian(n):
    X = np.random.normal(0,1,n)*np.sqrt(0.1)
    X = np.cumsum(X)
    return X
```

```python
plt.figure(figsize=(18,10))
plt.xlim(0,1000)
plt.ylim(0,500000)
for _ in range(50):
    sim = 300*np.exp(brownian(1000))
    plt.subplot(1,2,1)
    plt.plot(sim)
    plt.subplot(1,2,2)
    plt.plot(np.log(sim))
```

```python
# random graphs demonstrating percolation critical probability

def percolation(n,p):
    Z = np.random.binomial(1,p,(n,n))
    return Z

plt.figure(figsize=(18,10))
plt.subplot(1,4,1)
plt.imshow(percolation(100,0.3), cmap='gray')
plt.subplot(1,4,2)
plt.imshow(percolation(100,0.5), cmap='gray')
plt.subplot(1,4,3)
plt.imshow(percolation(100,0.7), cmap='gray')
plt.subplot(1,4,4)
plt.imshow(percolation(100,0.1), cmap='gray')
```

```python
import networkx as nx

def graph(n,p):
    Z = np.random.binomial(1,p,(n,n))
    G = nx.from_numpy_array(Z)
    return G

for _ in range(10):
    plt.figure(figsize=(18,10))
    plt.subplot(2,4,1)
    nx.draw(graph(6,0.1))
    plt.subplot(2,4,2)
    nx.draw(graph(6,0.2))
    plt.subplot(2,4,3)
    nx.draw(graph(6,0.3))
    plt.subplot(2,4,4)
    nx.draw(graph(6,0.4))
    plt.subplot(2,4,5)
    nx.draw(graph(6,0.5))
    plt.subplot(2,4,6)
    nx.draw(graph(6,0.6))
    plt.subplot(2,4,7)
    nx.draw(graph(6,0.7))
    plt.subplot(2,4,8)
    nx.draw(graph(6,0.8))
```
