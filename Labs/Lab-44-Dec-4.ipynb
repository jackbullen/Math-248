```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import solve_ivp
```

```python
# Lorenz System
s, b, r = 10, 8/3, 28
system = lambda t,x: [s*(x[1]-x[0]),x[0]*(r-x[2])-x[1],x[0]*x[1]-b*x[2]]
sol = solve_ivp(system, [0,10], [1,1,1])

sol
```

```python
for var in sol.y:
    plt.plot(sol.t, var)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(*sol.y)
```
